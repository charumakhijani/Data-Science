```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('sine wave form')
```

```python
x=np.arange(0, 5*np.pi, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

plt.subplot(2,1,1)
plt.plot(x, y_sin,'g--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x, y_cos,'r--')
plt.title('Cosine')
```

# Bar plot

```python
x=[2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]

plt.bar(x,y)
plt.bar(x2,y2)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('bar graph')
```

# Box Plot

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
data
```

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
plt.boxplot(data, vert=False, patch_artist=False)
```

# Pie Chart

```python
labels = 'Python', 'R', 'Julia', 'Scala'
sizes = [215,130,245,210]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.1, 0 ,0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
        
plt.axis('equal')
```
